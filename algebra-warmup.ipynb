# Noktaları Çizme

Bugünün ilk görevleri:
1. Bir nokta çiz
1. Birden fazla nokta çiz

Bağlantısız noktaları çizmek için [`plt.scatter()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter.) kullanacağız.

Birçok argüman alabilir, ancak ilk iki pozisyonel argüman en önemlisidir: `x` ve `y`.

## Bir Nokta Çiz

**x=5** ve **y=10** konumlarında bir nokta çiz

```python
import matplotlib.pyplot as plt

dot_coordinates = [5, 10]

plt.scatter(dot_coordinates[0], dot_coordinates[1])
plt.show()
```

## Birden Fazla Nokta Çiz

Bir listede saklanan birkaç noktayı **boyutlar (2, 13)** (yani bir listenin listesi) ile çiz.

İpucu: `x` ve `y` argümanları x'lerin ve y'lerin listeleri olabilir.

```python
import matplotlib.pyplot as plt

dots = [
    [1, 1, 2, 2, 2, 3, 1, 3, 3, 3, 3, 0, 0],
    [2, 3, 2, 4, 1, 3, 4, 2, 0, 5, 7, 2, 3]
]

plt.scatter(dots[0], dots[1])
plt.show()
```

Bu durumda x ve y için listeleri kullandık, ancak bu NumPy 1D-dizileri veya Pandas Serileri ile de eşit derecede iyi çalışır.

# Bir Çizgi Çiz

**x=4, y=0** konumunda başlayan ve **x=20, y=4** konumunda biten bir çizgi çiz. Kullanman gereken fonksiyon [pyplot.plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot).

`plt.plot()` sana verdiği **noktaları** çizer ve **noktalar arasında düz bir çizgi** çizer. Yani, bir çizgi çizmek için sadece iki nokta sağlamanız gerekir ve `plt.plot()` bunların arasında çizgiyi çizecektir.

`plt.scatter()` gibi, `plt.plot()` iki argüman bekler: x'lerin bir listesi ve y'lerin bir listesi.

```python
import matplotlib.pyplot as plt

start = {"x": 4, "y": 0}
end = {"x": 20, "y": 4}

plt.plot(
    [start["x"], end["x"]],
    [start["y"], end["y"]]
)

plt.show()
```

## Fonksiyonları Çizme

Matematiksel bir fonksiyonu çizmek için her zaman bu tarifi izleyeceğiz:

1. Çizmek istediğiniz x-değerlerinin aralığını tanımlayın
1. Bu aralık içinde eşit aralıklı x-değerlerinin bir listesini veya dizisini oluşturun
1. Matematiksel fonksiyonunuzu kullanarak bu x-değerlerine karşılık gelen y-değerlerini hesaplayın
1. x- ve y-değerlerini çizin

### Bir Fonksiyonu Çiz

❓ Aralık `[-10, 10]` içinde veri noktaları oluştur. Bu noktalar senin `x` noktalarındır. Aşağıdaki fonksiyonu çizmek için bunları kullan:

$$y = 4x + 2$$

**Y-ekseni** **-10 ile 10** arasında tutmak için [ plt.ylim(-10, 10)](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylim.html#matplotlib.pyplot.ylim) kullan

<br/>

<details>
  <summary>💡 İpucu</summary>
  <code><a href="https://docs.scipy.org/doc/numpy/reference/generated/numpy.arange.html">numpy.arange()</a></code> veya <code><a href="https://numpy.org/doc/stable/reference/generated/numpy.linspace.html">numpy.linspace()</a></code> kullanarak bir aralık içinde N sayı listesi oluşturabilirsin.

  Fark ne?
  - Her ikisi de başlangıç ve bitiş değeri alır
  - `np.arange()` ile aralar için adım boyutunu tanımlarsın
  - `np.linspace()` ile aralar için nokta sayısını tanımlarsın
</details>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y = 4 * x + 2

plt.plot(x, y)

plt.ylim(-10, 10)

plt.show()
```

### Daha Gelişmiş Fonksiyonları Çiz

❓ Bu 4 fonksiyonu yeni bir grafikte birlikte çiz, yukarıda oluşturulan aynı `x` noktalarını kullanarak, **y-ekseni** **-2 ile 6** arasında ve **x-ekseni** **-4 ile 4** arasında tutarak.

$$y1 = x^2$$

$$y2 = e^x$$

$$y3 = \frac{1}{1 + e^{-(-3 + 2*x)}}$$

<details>
  <summary>💡 İpuçları</summary>
    
  Bunu yapmanın birden fazla yolu vardır... birkaç tanesini adlandırmak gerekirse:
  - Bir `for` döngüsü/liste anlayışı kullan
  - NumPy vektörel fonksiyonlarını kullan (örn. `np.exp(your_list_of_x_numbers)`)
  - Python yerleşik [`map`](https://www.w3schools.com/python/ref_func_map.asp) fonksiyonunu kullan

  Pro ipucu: NumPy vektörel fonksiyonlarını kullan: fonksiyonu sanki matematik yazıyormuş gibi yaz. Sadece daha hızlı değil, aynı zamanda okunması da daha kolay!

  `np.exp()` gerekli olacak.

</details>

Çok açılı bir çizgi mi alıyorsun? Daha küçük bir adım boyutuyla dene. `plt.plot()` sadece verdiğin x'leri ve y'leri çizer ve bunların arasında **düz** çizgi çizer. Noktalarınız birbirinden ne kadar yakın olursa, çizginiz o kadar düzgün olur.

```python
import numpy as np
import matplotlib.pyplot as plt

# Daha düzgün eğriler için çok nokta
x = np.linspace(-4, 4, 1000)

y1 = x**2
y2 = np.exp(x)
y3 = 1 / (1 + np.exp(-(-3 + 2*x)))
y4 = np.sin(x)

plt.plot(x, y1, label="x²")
plt.plot(x, y2, label="e^x")
plt.plot(x, y3, label="sigmoid")
plt.plot(x, y4, label="sin(x)")

plt.xlim(-4, 4)
plt.ylim(-2, 6)

plt.legend()
plt.grid(True)

plt.show()
```

## Noktaları ve Çizgileri Birleştir

Aşağıda `sklearn` ile oluşturulmuş **dağılım grafiği** var, sizin yakında öğreneceğiniz bir kütüphane! Noktaların dağılımını mümkün olduğunca yakından takip eden **kırmızı noktalı düz bir çizgi** çizmek istiyoruz 🖌️

Son sonucumuz şöyle görünmelidir:  
<img src="https://wagon-public-datasets.s3.amazonaws.com/data-science-images/02-Data-Toolkit/03-Data-Visualization/020302-regression-line-example.png" alt="regression_line" width=400>

```python
from sklearn import datasets

x, y, _ = datasets.make_regression(n_samples=100, n_features=1, noise=10, coef=True, random_state=0)

plt.scatter(x,y);
```

❓ Karmaşık hesaplamalara gerek yok, sadece gözlerinizi kullanarak çizginin ne olması gerektiğini tahmin et.

Yukarıdaki dağılım grafiğini **yaklaşık olarak** regrese eden kırmızı noktalı bir çizgi çiz
- Önceki hücreden oluşturulan `x` ve `y` yi yeniden kullan
- Hatırla: `plt.plot()` ile bir çizgi çizmek için iki nokta yeterlidir
- Çizginizi kırmızı ve noktalı yapmak için şu argümanları kullan: `color='r', linestyle='--'` <br>(`color` için diğer değerlerle dene: `'g'` veya `'c'` veya `'m'` ve `linestyle` için: `':'` veya `'-.'`)

```python
plt.plot(
    [-2, 2],
    [-90, 90],
    color='r',
    linestyle=':'
)

plt.show()
```

❓ Şimdi, grafiğimize bağlam ekleyeceğiz:
1. "Doğrusal Regresyon" diyen bir başlık
1. Veri noktalarını ve çizgiyi açıklayan bir açıklama
1. X-ekseni için etiket ("giriş değeri") ve y-ekseni için etiket ("çıkış değeri")

Kodu buraya sağlıyoruz. Sadece `# YOUR CODE HERE` yi daha önce oluşturduğun grafikle değiştir, ancak bu sefer iki grafiğe de bir `label` ekle.

```python
fig = plt.figure(figsize=(10, 5))

plt.scatter(x, y, label="Generated data")

plt.plot(
    [-2, 2],
    [-90, 90],
    color='r',
    linestyle=':',
    label="Regression line"
)

plt.title("Linear Regression")
plt.legend()

plt.xlabel("input value")
plt.ylabel("output value")

plt.annotate("Regression line", (0,0), (1,-50),
             arrowprops={"arrowstyle":"->"});
```

```python
from sklearn import datasets
import matplotlib.pyplot as plt

x, y, _ = datasets.make_regression(
    n_samples=100,
    n_features=1,
    noise=10,
    coef=True,
    random_state=0
)

x = x.flatten()

fig = plt.figure(figsize=(10, 5))

plt.scatter(x, y, label="Generated data")

plt.plot(
    [-2, 2],
    [-90, 90],
    color="r",
    linestyle=":",
    label="Regression line"
)

plt.title("Linear Regression")
plt.legend()

plt.xlabel("input value")
plt.ylabel("output value")

plt.annotate("Regression line", (0, 0), (1, -50),
             arrowprops={"arrowstyle": "->"});
```
